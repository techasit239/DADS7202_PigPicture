```python
import os
from PIL import Image
from google.colab import drive
from pillow_heif import register_heif_opener

# а№Ђаё›аёґаё”а№ѓаё«а№‰ PIL аё­а№€аёІаё™а№„аёџаёҐа№Њ HEIC / HEIF а№„аё”а№‰
register_heif_opener()

# а№ЂаёЉаё·а№€аё­аёЎаё•а№€аё­ Google Drive
drive.mount('/content/drive')

# а№‚аёџаёҐа№Ђаё”аё­аёЈа№Њаё•а№‰аё™аё—аёІаё‡аё«аёҐаё±аёЃ
SOURCE_FOLDER = '/content/drive/MyDrive/FreshCheck/data/Int spoiled pork'

# а№‚аёџаёҐа№Ђаё”аё­аёЈа№Њаё›аёҐаёІаёўаё—аёІаё‡
DESTINATION_FOLDER = '/content/drive/MyDrive/FreshCheck/data/Int spoiled pork_JPG'


def make_unique_path(path):
    if not os.path.exists(path):
        return path

    base, ext = os.path.splitext(path)
    counter = 1

    while True:
        new_path = f"{base}_{counter}{ext}"
        if not os.path.exists(new_path):
            return new_path
        counter += 1


def convert_all_images_to_jpg(source_root, destination_root):
    supported_formats = (
        '.png', '.jpeg', '.jpg', '.webp',
        '.bmp', '.tiff', '.tif', '.gif',
        '.heic', '.heif' ,'.HEIC'
    )

    total_success = 0
    total_error = 0
    skipped = 0

    for root, dirs, files in os.walk(source_root):
        rel_path = os.path.relpath(root, source_root)

        if rel_path == ".":
            target_dir = destination_root
        else:
            target_dir = os.path.join(destination_root, rel_path)

        os.makedirs(target_dir, exist_ok=True)

        for filename in files:
            if not filename.lower().endswith(supported_formats):
                skipped += 1
                continue

            source_path = os.path.join(root, filename)

            name_without_ext = os.path.splitext(filename)[0]
            destination_path = os.path.join(target_dir, f"{name_without_ext}.jpg")
            destination_path = make_unique_path(destination_path)

            try:
                with Image.open(source_path) as img:
                    # аё–а№‰аёІа№Ђаё›а№‡аё™ GIF а№ѓаёЉа№‰ frame а№ЃаёЈаёЃ
                    if getattr(img, "is_animated", False):
                        img.seek(0)

                    # аё–а№‰аёІаё аёІаёћаёЎаёµаёћаё·а№‰аё™аё«аёҐаё±аё‡а№‚аё›аёЈа№€аё‡а№ѓаёЄ а№ѓаё«а№‰а№ѓаёЄа№€аёћаё·а№‰аё™аё«аёҐаё±аё‡аё‚аёІаё§аёЃа№€аё­аё™а№Ѓаё›аёҐаё‡а№Ђаё›а№‡аё™ JPG
                    if img.mode in ('RGBA', 'LA') or (
                        img.mode == 'P' and 'transparency' in img.info
                    ):
                        img = img.convert('RGBA')
                        background = Image.new('RGB', img.size, (255, 255, 255))
                        background.paste(img, mask=img.split()[-1])
                        rgb_img = background
                    else:
                        rgb_img = img.convert('RGB')

                    rgb_img.save(destination_path, 'JPEG', quality=95)

                print(f"аёЄаёіа№ЂаёЈа№‡аё€: {source_path} -> {destination_path}")
                total_success += 1

            except Exception as e:
                print(f"а№Ѓаё›аёҐаё‡а№„аёЎа№€а№„аё”а№‰: {source_path} | Error: {e}")
                total_error += 1

    print("\n==============================")
    print("аёЄаёЈаёёаё›аёњаёҐаёЃаёІаёЈа№Ѓаё›аёҐаё‡а№„аёџаёҐа№Њ")
    print("==============================")
    print(f"а№Ѓаё›аёҐаё‡аёЄаёіа№ЂаёЈа№‡аё€: {total_success} а№„аёџаёҐа№Њ")
    print(f"а№Ѓаё›аёҐаё‡а№„аёЎа№€аёЄаёіа№ЂаёЈа№‡аё€: {total_error} а№„аёџаёҐа№Њ")
    print(f"аё‚а№‰аёІаёЎа№„аёџаёҐа№Њаё—аёµа№€а№„аёЎа№€а№ѓаёЉа№€аёЈаё№аё›аё аёІаёћ: {skipped} а№„аёџаёҐа№Њ")
    print(f"аёљаё±аё™аё—аё¶аёЃа№„аё§а№‰аё—аёµа№€: {destination_root}")


print("--- а№ЂаёЈаёґа№€аёЎаё•а№‰аё™а№Ѓаё›аёҐаё‡а№„аёџаёҐа№Њаё—аё±а№‰аё‡аё«аёЎаё” ---")
convert_all_images_to_jpg(SOURCE_FOLDER, DESTINATION_FOLDER)
print("--- а№ЂаёЄаёЈа№‡аё€аёЄаёґа№‰аё™ ---")
```
